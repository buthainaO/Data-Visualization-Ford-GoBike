## Dataset Overview

> The dataset for Ford GoBike aggregates the data for rides in February 2019, Ford GoBike is the Bay Area’s bike share system. Bay Area Bike Share was introduced in 2013. It services San Francisco, the East Bay and San Jose. It contains 183412
 row and 16 columns.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
#Loading the dataset
df_clean = pd.read_csv('clean-fordgobike-tripdata.csv')
```

## (Frequency by weekdays)

> The plot shows the distribution between weekdays. The distribution between weekdays shows that Thursday and Tuesday are the peak days and most of the rides are during these days, on the other hand Sunday and Saturday are the least days. 
We can conclude from this information that most of the members are using it for work so there’s no need to use it on the weekends. 

```python
base_color = sb.color_palette()[0]
plt.figure(figsize=[ 11.69, 8.27])
order_days = ['Mon', 'Tue','Wed','Thu','Fri','Sat','Sun']
sb.countplot(data=df_clean, x='start_weekday_char', color=base_color, order = order_days)
plt.title('Frequency by weekdays', fontsize = 18)
plt.xlabel('Weekdays')
plt.ylabel('Counts');
```

## (Durations of rides for age groups)

> This plot shows the duration of rides for each age group. The age group 30-40 tends to have more duration for riders. 

```python
plt.figure(figsize=[ 11.69, 8.27])
plt.scatter(data = df_clean, x = 'member_age', y = 'duration_min', alpha = 0.2);
plt.style.use('seaborn')
plt.title('Durations of rides for age groups', fontsize = 18)
plt.xlabel('Age')
plt.ylabel('Duration by minutes')
plt.xlim(df_clean['member_age'].min(), 100);
```

## (Comparing ride duration across ages and genders)

> This plot compares the duration of rides across ages and user types. The subscriber type tends to overlap in small durations and the customer type has more durations. Most of the two types are between 20-60 years.

```python
cat_markers = [['Customer', 's'],
               ['Subscriber', '^']]
plt.figure(figsize=[ 11.69, 8.27])
for cat, marker in cat_markers:
    df_cat = df_clean[df_clean['user_type'] == cat]
    plt.scatter(data = df_cat, x = 'member_age', y = 'duration_sec', marker = marker, alpha = .5);
plt.legend(['Customer', 'Subscriber']);
plt.xlim(df_clean['member_age'].min(), 100);
plt.title('Comparing Ride Duration Across Ages and Genders', fontsize = 18);
plt.xlabel('Ages');
plt.ylabel('Number of Rides');
plt.style.use('seaborn')
```
